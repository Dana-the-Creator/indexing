```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
from collections import defaultdict

# TF-IDF 벡터화
vectorizer = TfidfVectorizer(stop_words="english")  # 불용어 제외
X = vectorizer.fit_transform(pages_text)

# TF-IDF 점수가 높은 단어 추출
feature_names = np.array(vectorizer.get_feature_names_out())
sorted_indices = np.argsort(X.sum(axis=0).A1)[::-1]  # 높은 TF-IDF 점수 순으로 정렬

# 상위 50개의 중요한 단어 추출
top_n = 50
top_words = [feature_names[idx] for idx in sorted_indices[:top_n]]

# 색인 생성: 각 단어가 등장한 페이지 추적
index = defaultdict(list)

for page_num, page_text in enumerate(pages_text):
    for word in top_words:
        if word in page_text:
            index[word].append(page_num + 1)  # 페이지 번호는 1부터 시작

# 색인 출력
for word, pages in sorted(index.items()):
    print(f"단어: {word}, 페이지: {pages}")
```

```python
from konlpy.tag import Okt
from collections import defaultdict

# Okt 객체 생성
okt = Okt()

# 명사와 동사 추출
def extract_nouns_and_verbs(text):
    # 형태소 분석을 통해 명사와 동사 추출
    morphs = okt.pos(text)
    nouns_and_verbs = [word for word, tag in morphs if tag in ['Noun', 'Verb']]
    return nouns_and_verbs

# 새로운 색인 생성
def generate_noun_verb_index(pages_text):
    index = defaultdict(list)  # 단어: 페이지 번호 리스트
    for page_num, text in enumerate(pages_text, 1):
        words = extract_nouns_and_verbs(text)
        for word in words:
            index[word].append(page_num)
    return index

# 명사와 동사로 색인 생성
noun_verb_index = generate_noun_verb_index(pages_text)

# 색인 출력
for word, pages in noun_verb_index.items():
    print(f"단어: {word}, 페이지: {pages}")

```

```python
import gensim
from gensim import corpora
from konlpy.tag import Okt

# Okt 객체 생성
okt = Okt()

# LDA 모델 훈련 (여기서는 예시로 3개의 주제를 사용)
# 페이지 텍스트 전처리: 명사 추출
def preprocess(text):
    nouns = okt.nouns(text)
    return nouns

# 전처리된 텍스트
processed_text = [preprocess(text) for text in pages_text]

# 사전 생성 (Dictionary 생성)
dictionary = corpora.Dictionary(processed_text)

# 문서-단어 매트릭스 생성 (문서에 단어의 빈도)
corpus = [dictionary.doc2bow(text) for text in processed_text]

# LDA 모델 훈련
lda = gensim.models.LdaModel(corpus, num_topics=3, id2word=dictionary, passes=15)

# 주제별 상위 단어 추출
topics = lda.print_topics(num_words=10)  # 각 주제별 상위 10개 단어

# 주제에서 추출된 단어들만 사용
topic_words = []
for idx, topic in topics:
    words = topic.split(" + ")
    for word in words:
        word = word.split("*")[1].strip().replace('"', '')  # 단어만 추출
        topic_words.append(word)

# 페이지별 단어 추출
page_word_index = {}

# 각 페이지 텍스트에서 주제에 관련된 단어 추출
for page_num, text in enumerate(pages_text, start=1):
    processed_text = preprocess(text)
    
    # 해당 페이지에서 나온 단어들 중 주제에 포함된 단어만 추출
    for word in processed_text:
        if word in topic_words:
            if word not in page_word_index:
                page_word_index[word] = set()  # set 사용하여 중복 제거
            page_word_index[word].add(page_num)

# 결과 출력 (단어: 페이지 번호)
for word, page_nums in page_word_index.items():
    print(f"{word}: 페이지 {', '.join(map(str, page_nums))}")
```
